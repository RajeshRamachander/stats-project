```python
# Central Limit Theorem

def central_limit_theorem(data,n_samples = 1000, sample_size = 500):
    """ Use this function to demonstrate Central Limit Theorem. 
        data = 1D array, or a pd.Series
        n_samples = number of samples to be created
        sample_size = size of the individual sample """
    %matplotlib inline
    import pandas as pd
    import numpy as np
    import matplotlib.pyplot as plt
    import seaborn as sns
    min_value = 0  # minimum index of the data
    max_value = data.count()  # maximum index of the data
    b = {}
    for i in range(n_samples):
        x = np.unique(np.random.randint(min_value, max_value, size = sample_size)) # set of random numbers with a specific size
        b[i] = data[x].mean()   # mean of each sample
    c = pd.DataFrame()
    c['sample'] = b.keys()  # sample number 
    c['Mean'] = b.values()  # mean of that particular sample
    plt.figure(figsize= (15,5))

    plt.subplot(1,2,2)
    sns.distplot(c.Mean)
    plt.title(f"Sampling Distribution. \n \u03bc = {round(c.Mean.mean(), 3)} & SE = {round(c.Mean.std(),3)}")
    plt.xlabel('data')
    plt.ylabel('freq')

    plt.subplot(1,2,1)
    sns.distplot(data)
    plt.title(f"Population Distribution. \n \u03bc = {round(data.mean(), 3)} & \u03C3 = {round(data.std(),3)}")
    plt.xlabel('data')
    plt.ylabel('freq')

    plt.show()
```

#Calculating value of Pi = 3.14

1. Initialize the number of (x,y) cordiantes in circle and square to zero. Also define the number of design points 
2. Generate random point x.
3. Generate random point y.
4. Calculate the euclidean distance from origin to (x,y) as d = x*x + y*y.# what we know

5. If d <= 1, increment c_points.
6. Always increment the s_points.
7. Repeat till the if condition is met. # simulate the process

8. Calculate pi = 4*(c_points/s_points). # answer

9. Terminate.


```python
import random 
  
samplespace= 100000 # step 1
  
c_points= 0 # circle points
s_points= 0 # square points
x=[]
y=[]
  
# Total Random numbers generated is combination of (x,y) pair
# step 1 : finding the random space
for i in range(samplespace): 
  
    # First we generate random values of (x,y) cordinates to find the distance from uniform distribution 
    r_x= random.uniform(-1, 1) # find the prob space
    r_y= random.uniform(-1, 1) 

    x.append(r_x) # storing the x coordinate values to a variable
    y.append(r_y) # storing the y coordinate values to a variable
  
    # Next calculate the Euclidean distance between (x, y) from the origin 
    euclidiean_dist= np.sqrt(r_x**2 + r_y**2) # distance from origin to (x,y)
  
    # Checking if (x, y) lies inside the circle 
    if euclidiean_dist<= 1: 
        c_points = c_points + 1 # update the c_points
  
    s_points = s_points + 1 # update the s_points
  
    
    area_ratio = c_points/ s_points 
    
    pi = 4*area_ratio
    
    pi_lower = pi - 4*1.96*np.sqrt(area_ratio*(1-area_ratio)/samplespace)
    pi_upper = pi + 4*1.96*np.sqrt(area_ratio*(1-area_ratio)/samplespace)
 
  
print("Estimate of Pi=", pi)  
print("95% confidence interval for Pi is", pi_lower, "to",pi_upper)
```

#Monty hall problem

```python
import numpy as np
import matplotlib.pyplot as plt
from time import time as TT
from random import choice as ch
import numpy as np
from random import randint
```

We will simulate the game of monty hall problem here. The key assumptions are (refer to google doc for detailed explanation) :
- Host knows the ground truth
- The player will choose a door to open # aman

- Host will eliminate one door based on players choice # door 1
- Player WILL change his choice accordingly
- Players chance of winning will be near to ~66.7 % (From Bayes theorem)
- For each iteration the number of times the games are played will increase from 1 to N (100,200,300...,10000)



- Maximum iteration = 1000
- doors = [1,2,3]
- rep hold value of ground truth and players choice
- hx and px refers to host and playres choices respectively
- score = 1 or 0 based on alternative decision made by the player. 1 if alternatice decision matches the ground truth
- mean_score = holds the mean value of score from each iteration

```python
#step 1 : generate random number
#step 2 : 

ac=[]
N=[]
max_iteration = 100 #consumes RAM. Be careful here.
for z in range(1,max_iteration): #Simulating the game
  doors = [1,2,3] #we have three doors here
  score = [] 
  for k in range(1,z):
    rep = [] #variable
    player = doors.copy()
    rand1 = randint(0,2) # 0 ,1 ,2 
    hx = doors[rand1] #ground truth choosen by host
    rep.append(hx)
    rand2= randint(0,2) #players choice
    px=doors[rand2]
    rep.append(px) #adding the choices to rep variavble
    eliminate = [] #the host should decide which door to eliminate ?
    for i in doors:
      if i not in rep:
        eliminate = i  #eliminate the door where no prize and not the choice made by Aman

    #print(" The eliminated door number is ", eliminated)

    player.remove(eliminate) #remove the door opened by host as an alternative choice
    player.remove(px)    #remove the original player choice
    alternate_choice = player[0] # the only remaining option

    #check if the alternate selection matched with ground truth

    if alternate_choice==hx:
      score.append(1)
    else:
      score.append(0)

    games_played = k

  ac.append(np.mean(score))
  N.append(z)

game_results = np.delete(ac, 0) #first value is NaN. So remove it.
mean_win_percentage = np.mean(game_results)
print ("As per Bayes theorem win percentage if player switch the door= 66.7%")
print("The win percentage from random simulation is near to %f precentage." % (round(mean_win_percentage*100,2)))
```

```python
import numpy as np
import scipy.stats as st
import seaborn as sns
```

```python
#Acceptance rejection definition
# Supply-demand G(.)=L-R or S-D
def circle(x, y):
    return (x-1)**2 + (y-2)**2 - 3**2
```

```python
def mcmc(p, iter=1000): #MC
    x, y = 0., 0. #starting state
    #p(x,y) is the last known state
    #p(xstar, ystar) new state
    #we check weather the new state in acceptance-rejection region  0 to 1
    #match the population parameters (0.95). Reliable sample.

    samples = np.zeros((iter, 2))

    for i in range(iter): #montecarlo
        x_star, y_star = np.array([x, y]) + np.random.normal(size=2) #markov chain
        if np.random.rand() < p(x_star, y_star) / p(x, y):
            x, y = x_star, y_star #update accepted as point in new distibution #MCMC
        samples[i] = np.array([x, y]) #updating the target sample space

    return samples  
```

```python
samples = mcmc(circle, iter=1000)
sns.jointplot(samples[:, 0], samples[:, 1])

#samples = metropolis_hastings(pgauss, iter=10000)
#sns.jointplot(samples[:, 0], samples[:, 1])
```
